MatplotLib is a plotting library for python programming language and its mathematical extension NumPy. It provides an Object-Oriented API for embedding plots into applications using general purpose GUI toolkits like Tinkter, Qt, wxPython etc. 

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#using %matplotlib inline--> do not need to write plt.show , it will automatically give the plot
```

```python
import numpy as np
```

```python
x= np.arange(0,10)
y=np.arange(11,21)
```

```python
##plt scatter 
plt.scatter(x,y,c='g')
#c-->color 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')

#to save this figure, run the following code snippet 
#plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y)

#if i want a red color line with points 
#'r*-'
plt.plot(x,y,'r*-')
#similarly 
#'ro-'
plt.plot(x,y,'ro-')
```

```python
#if you want to create subplots 
#that is compare different plots 
plt.subplot(2,2,1)
#first argument--> number of rows
#second argument--> mumber of columns
#third argument--> index 
#index--> konsi position pe konsa plot aayega 

plt.plot(x,y,'r*-')
plt.subplot(2,2,2)
plt.plot(x,y,'go-')
plt.subplot(2,2,3)
plt.plot(x,y,'m--')

#atmost 4 plots can be drawn , since the size of the matrix is 2 rows* 2 columns
```

```python
x= np.arange(0,10)
y= 3*x + 5 
plt.plot(x,y,'o')
#if you do not mention the color, the default color is blue 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('MatplotLib Demo')
```

```python
#simlarly different sine and cosine waves can be formed 
x= np.arange(0, 4*np.pi, 0.1)
y= np.sin(x)
plt.plot(x,y,'go')
```

```python
y_sin= np.sin(x)
y_cos= np.cos(x)
y_tan= np.tan(x)
plt.subplot(3,1,1)
plt.plot(x, y_sin)
plt.subplot(3,1,2)
plt.plot(x,y_cos)
plt.subplot(3,1,3)
plt.plot(x, y_tan)
```

```python
##Bar Plot 
x1= [2,8,10]
y1= [4, 16 ,20]
x2=[3,9,11]
y2=[6, 18,22]
plt.bar(x1,y1)
plt.bar(x2,y2,color='red')
```

```python
##Histograms

a= np.array([22,1,2,3,6,7,34,23,56,78,65,89,90,95,98,99,75,77,79])
plt.hist(a, bins=20)

#by default the value of bins-->10 
#this implies that, it divided the the range from 0-10, 10-20 
#in the Y-axis, we get, how many numbers in the array lie between 0-10 , 10-20, 20-30 etc 
```
